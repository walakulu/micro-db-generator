#### Scrips For Lookup Tables

```python
#  Generate Transaction Type (LOOKUP table)

def transaction_type_map(id,txn_type):
    
    SQL=f"INSERT INTO `Transaction_Type` (`id`, `type` ) VALUES ('{id}', '{txn_type}');\n"
    output_file_name=f"sql/OLTP/experiment-1/transaction_type.sql"
    
    with open(output_file_name, 'a') as f:
        f.write(SQL)

    return SQL

txn_types=transactions_df.tx_type.value_counts().index.sort_values().tolist()

for txn_type in txn_types: 
    txn_type_id=txn_types.index(txn_type)
    transaction_type_insert_sql=transaction_type_map(txn_type_id,txn_type)
```
