### Generate SQL Scripts For Temporal Databases

#### Scrips For Temporal Tables

```python

txn_types=transactions_df.tx_type.value_counts().index.sort_values().tolist()

# Generate Customer Profiles

def customer_profile_map(txn_id,txn_timestamp,temporal_priod,account_df):
    
    cus_id = account_df['dsply_nm'].iloc[0]
    ssn = account_df['ssn'].iloc[0]
    first_name = account_df['first_name'].iloc[0]
    last_name = account_df['last_name'].iloc[0]
    street_address = account_df['street_addr'].iloc[0]
    city= account_df['city'].iloc[0]
    state = account_df['state'].iloc[0]
    country = account_df['country'].iloc[0]
    zip = account_df['zip'].iloc[0]
    gender = account_df['gender'].iloc[0]
    birth_date = account_df['birth_date'].iloc[0]
    txn_id = txn_id
    valid_start_time = txn_timestamp
    valid_end_time = valid_start_time + timedelta(days=temporal_priod)
    
    SQL=f"INSERT INTO `Customer` (`cus_id`, `ssn`, `first_name`, `last_name`, `street_address`, `city`, `state`, `country`, `zip`, `gender`, `birth_date`, `txn_id`, `valid_start_time`, `valid_end_time`) VALUES ('{cus_id}', '{ssn}', '{first_name}', '{last_name}', '{street_address}', '{city}', '{state}', '{country}', '{zip}', '{gender}', '{birth_date}', '{txn_id}', '{valid_start_time}', '{valid_end_time}');\n"
    
    output_file_name=f"sql/TEMPORAL/customer_profile_temporal_{temporal_priod}_day.sql"
    
    with open(output_file_name, 'a') as f:
        f.write(SQL)

    return SQL

#  Generate Wallets

def wallet_map(txn_id,txn_timestamp,temporal_priod,account_df):
    wallet_id = account_df['acct_id'].iloc[0]
    wallet_type = account_df['type'].iloc[0]
    acct_stat = account_df['acct_stat'].iloc[0]
    currency_type = account_df['acct_rptng_crncy'].iloc[0]
    prior_sar_acc= 1 if (account_df['prior_sar_count'].iloc[0] == 'True') else 0
    open_dt = account_df['open_dt'].iloc[0]
    close_dt = account_df['close_dt'].iloc[0]
    initial_deposit = account_df['initial_deposit'].iloc[0]
    cus_id = account_df['dsply_nm'].iloc[0]
    txn_id = txn_id
    valid_start_time = txn_timestamp
    valid_end_time = valid_start_time + timedelta(days=temporal_priod)
    
    SQL=f"INSERT INTO `Wallet` (`wallet_id`, `type`, `wallet_status`, `currency_type`, `prior_sar_acc`, `open_dt`, `close_dt`, `initial_deposit`, `cus_id`, `txn_id`, `valid_start_time`, `valid_end_time`) VALUES ('{wallet_id}', '{wallet_type}', '{acct_stat}', '{currency_type}', '{prior_sar_acc}', '{open_dt}', '{close_dt}', '{initial_deposit}', '{cus_id}', '{txn_id}', '{valid_start_time}', '{valid_end_time}');\n"
    
    output_file_name=f"sql/TEMPORAL/wallet_temporal_{temporal_priod}_day.sql"
    
    with open(output_file_name, 'a') as f:
        f.write(SQL)

    return SQL

#  Generate Transactions
def transaction_map(temporal_priod,transaction_record):
    txn_id=row['tran_id']
    txn_amt=row['base_amt']
    source_acc=row['orig_acct']
    destination_acc=row['bene_acct']
    txn_timestamp=row['tran_timestamp']
    txn_type_id=txn_types.index(row['tx_type'])
    valid_start_time = txn_timestamp
    valid_end_time = valid_start_time + timedelta(days=temporal_priod)
    
    SQL=f"INSERT INTO `Transaction` (`txn_id`, `txn_amt`, `orig_acc`, `bene_acc`, `txn_timestamp`, `txn_type_id`,`valid_start_time`, `valid_end_time`) VALUES ('{txn_id}', '{txn_amt}', '{source_acc}', '{destination_acc}', '{txn_timestamp}', '{txn_type_id}', '{valid_start_time}', '{valid_end_time}');\n"
    
    output_file_name=f"sql/TEMPORAL/transaction_temporal_{temporal_priod}_day.sql"
    
    with open(output_file_name, 'a') as f:
        f.write(SQL)

    return SQL    


for index, row in transactions_df.iterrows():
    # only for temporal tables   
    temporal_priod=28
    txn_id=row['tran_id']
    txn_timestamp=row['tran_timestamp']
    source_acc=row['orig_acct']
    destination_acc=row['bene_acct']
    source_profile_df=accounts_df[accounts_df.acct_id==source_acc]
    destination_profile_df=accounts_df[accounts_df.acct_id==destination_acc]
    
    #create profiles for source and destination accounts. 1 tnx associated with 2 accounts
    source_profile_insert_sql=customer_profile_map(txn_id,txn_timestamp,temporal_priod,source_profile_df)
    destination_profile_insert_sql=customer_profile_map(txn_id,txn_timestamp,temporal_priod,destination_profile_df)
    
    #create wallets for source and destination accounts.1 tnx associated with 2 waletes (ca be duplicated)
    wallet_account_insert_sql=wallet_map(txn_id,txn_timestamp,temporal_priod,source_profile_df)
    wallet_account_insert_sql=wallet_map(txn_id,txn_timestamp,temporal_priod,destination_profile_df)
    
    # create transactions
    transactio_insert_sql=transaction_map(temporal_priod,row)
        
```

#### Scrips For Lookup Tables

```python
#  Generate Transaction Type (LOOKUP table)
def transaction_type_map(id,txn_type,temporal_priod):
    
    SQL=f"INSERT INTO `Transaction_Type` (`id`, `type` ) VALUES ('{id}', '{txn_type}');\n"
    output_file_name=f"sql/TEMPORAL/transaction_type_temporal_{temporal_priod}_day.sql"
    
    with open(output_file_name, 'a') as f:
        f.write(SQL)

    return SQL

for txn_type in txn_types: 
    txn_type_id=txn_types.index(txn_type)
    transaction_type_insert_sql=transaction_type_map(txn_type_id,txn_type,temporal_priod)
```
